 A **for loop**  in Python is a control flow statement that is used to repeatedly execute a group of statements as long as the condition is satisfied. Such a type of statement is also known as an iterative statement. Thus, a for loop is an iterative statement.
<br>Example:

```python
from IPython.display import Image
Image("for-loop-700x234.png")
```

```python
plates = [1,2,3,4,5]
for x in plates:
 print('I put a cake slice on plate number', x)
print('All 5 plates are ready to be served!')
```

A **while loop** is a control flow statement which allows code to be executed repeatedly, depending on whether a condition is satisfied or not. As long as some condition is true, 'while' repeats everything inside the loop block. It stops executing the block if and only if the condition fails.

Let's assume you're writing a program for the washing cycle in your washing machine. Let's set the washing cycle to run 2 times. Thus, the following code will iterate 2 times using the 'while' loop:

```python
from IPython.display import Image
Image("while-loop.png")
```

```python
x = 1
print('I havent entered the loop yet')
print('Checked detergent, checked clothes')
while(x<=2):
     print('Hey there, I entered the loop')
     print('This is washing cycle number', x)
     x=x+1
print('Washing cycle complete. I exited the loop')
```
